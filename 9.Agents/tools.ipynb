Built-in Tool

DUCKDUCKGO Search

A tool that lets an LLM search the web using DuckDuckGo and retrieve real-world information.

LLMs don’t know current events or live data. This tool lets them:

Look up fresh information

Answer questions that require external knowledge

Avoid hallucinating facts.


```python
from langchain_community.tools import DuckDuckGoSearchRun
search_tool = DuckDuckGoSearchRun()
results = search_tool.invoke("top news in india today")
print(results)

```

```python
print(search_tool.name)
print(search_tool.description)
print(search_tool.args)
```

Custom Tools

1. it is used when you call your internal API.
2. A custom tool is required when your LLM needs to interact with your own logic, data, or systems in a way that existing tools don’t cover.


```python
from langchain_core.tools import tool

#  step 1- create a function 
def multiply(a, b):
    """Multiply two numbers"""
    return a*b
```

```python
# Step 2 - add type hints

def multiply(a:int , b:int) ->int: # hum btate hai llm ko in input and output kya hoga
    """Multiply two numbers"""
    return a*b
```

```python
# Step 3 - add tool decorator

@tool
def multiply(a: int, b:int) -> int:
    """Multiply two numbers"""
    return a*b
```

```python
result = multiply.invoke({"a":3 , "b":5})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

Method 2 - Using StructuredTool

it is used at production level.
it is more precise way to make tool.

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
def multiply_func(a: int, b: int) -> int:
    return a * b
```

```python
# making custom tool
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply", # tool name
    description="Multiply two numbers",
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

Toolkit

```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
    def get_tools(self): # it is a method
        return [add, multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
